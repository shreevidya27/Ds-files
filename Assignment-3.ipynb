```python
for col_name in data.select_dtypes(include=np.number).columns[:]:
    print(col_name)
    q1=data[col_name].quantile(.25)
    q3=data[col_name].quantile(.75)
    iqr=(q3-q1)
    low=q1-(1.5*iqr)
    high=q3+(1.5*iqr)
    print(data.loc[(data[col_name]<low) | (data[col_name]>high),col_name])
```
